```python
import fitz  # PyMuPDF
from docx import Document
import xml.etree.ElementTree as ET

def pdf_to_docx(pdf_file, docx_file):
    pdf_document = fitz.open(pdf_file)
    doc = Document()
    
    for page_num in range(pdf_document.page_count):
        page = pdf_document.load_page(page_num)
        
        # Create a new page in the DOCX file
        doc.add_page_break()
        
        for block in page.get_text("blocks"):
            text = ""
            for span in block[4]:
                text += span[4]
            
            style = block[0]
            
            if style["font"].startswith("Arial") and style["size"] == 14 and "bold" in style["flags"]:
                # If it's a 14-point Arial bold font, add it as a header
                doc.add_heading(text, level=1)
            else:
                # Otherwise, add it as regular text
                doc.add_paragraph(text)

    # Save the DOCX file
    doc.save(docx_file)

def docx_to_xml(docx_file, xml_file):
    doc = Document(docx_file)
    root = ET.Element("document")

    for paragraph in doc.paragraphs:
        if paragraph.style.name == 'Heading 1':
            # If it's a heading, create a header XML element
            element = ET.SubElement(root, "header")
        else:
            # Otherwise, create a text XML element
            element = ET.SubElement(root, "text")

        element.text = paragraph.text

    # Save the XML file
    tree = ET.ElementTree(root)
    tree.write(xml_file, encoding="utf-8", xml_declaration=True)

if __name__ == "__main__":
    pdf_file = "8.703-2020.pdf"   # Replace with your PDF file path
    docx_file = "output.docx"  # Output DOCX file path
    xml_file = "output.xml"    # Output XML file path

    # Convert PDF to DOCX while preserving styles
    pdf_to_docx(pdf_file, docx_file)
    
    # Extract headers from DOCX and save to XML
    docx_to_xml(docx_file, xml_file)
```

```python
import xml.etree.ElementTree as ET
from docx import Document
import re

# Helper function to convert DOCX paragraph styles to XML element attributes
def style_to_attributes(paragraph, element):
    run = paragraph.runs[0] if paragraph.runs else None

    # Font style
    if run and run.font.name:
        element.set("font", run.font.name)
    else:
        element.set("font", "Arial")  # Default font if not specified

    # Font size
    if run and run.font.size:
        element.set("font_size", str(run.font.size.pt))
    else:
        element.set("font_size", "12")  # Default font size if not specified

    # Bold, italic, and underline
    if run:
        element.set("bold", "true" if run.bold else "false")
        element.set("italic", "true" if run.italic else "false")
        element.set("underline", "true" if run.underline else "false")
    else:
        element.set("bold", "false")
        element.set("italic", "false")
        element.set("underline", "false")

def map_to_niso_sts_components(paragraph, element):
    # Map headings, numbered lists, links, titles, and sections to NISO STS components
    if paragraph.style.name.startswith("Heading"):
        element.tag = "title"
    elif re.match(r'^\d+\.', paragraph.text):
        runs = paragraph.runs
        if runs and runs[0].bold and runs[0].font.size and runs[0].font.size.pt == 12.0 and runs[0].font.name == "Arial":
            element.tag = "section"
    elif re.search(r'http|www', paragraph.text):
        element.tag = "ext-link"
        element.set("ext-link-type", "uri")
    elif element.tag != "p" and not paragraph.style.name.startswith("List"):
        element.tag = "title"

def docx_to_niso_sts_xml(docx_file, xml_file):
    doc = Document(docx_file)
    root = ET.Element("article")
    current_section = None

    for paragraph in doc.paragraphs:
        element = ET.Element("p")

        # Set style attributes
        style_to_attributes(paragraph, element)

        # Map to NISO STS components
        map_to_niso_sts_components(paragraph, element)

        # Extract and add the text content
        element.text = paragraph.text

        # Special handling for section IDs
        if element.tag == "section":
            section_id = element.text.split()[0]  # Extract the section ID (e.g., "10" from "10 Обеспечение качества измерений")
            if current_section is not None:
                # Close the previous section
                root.append(current_section)
            current_section = ET.Element("sec", {"id": f"s{section_id}"})
            label = ET.SubElement(current_section, "label")
            label.text = section_id

        # Add the element to the current section or the root
        if current_section is not None:
            current_section.append(element)
        else:
            root.append(element)

    # Close the last section, if any
    if current_section is not None:
        root.append(current_section)

    # Save the XML file
    tree = ET.ElementTree(root)
    tree.write(xml_file, encoding="utf-8", xml_declaration=True)

if __name__ == "__main__":
    docx_file = "output.docx"  # Replace with your Russian DOCX file path
    xml_file = "output.xml"                    # Output XML file path

    # Convert Russian DOCX to XML using NISO STS components, titles, and sections
    docx_to_niso_sts_xml(docx_file, xml_file)
```

```python
import xml.etree.ElementTree as ET
from docx import Document
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT
from docx.oxml.ns import nsdecls
from docx.oxml import parse_xml
import re

# Helper function to convert DOCX paragraph styles to XML element attributes
def style_to_attributes(paragraph, element):
    run = paragraph.runs[0] if paragraph.runs else None

    # Font style
    if run and run.font.name:
        element.set("font", run.font.name)
    else:
        element.set("font", "Arial")  # Default font if not specified

    # Font size
    if run and run.font.size:
        element.set("font_size", str(run.font.size.pt))
    else:
        element.set("font_size", "12")  # Default font size if not specified

    # Bold, italic, and underline
    if run:
        element.set("bold", "true" if run.bold else "false")
        element.set("italic", "true" if run.italic else "false")
        element.set("underline", "true" if run.underline else "false")
    else:
        element.set("bold", "false")
        element.set("italic", "false")
        element.set("underline", "false")

    # Typographic alignment
    if paragraph.alignment == WD_PARAGRAPH_ALIGNMENT.LEFT:
        element.set("alignment", "left")
    elif paragraph.alignment == WD_PARAGRAPH_ALIGNMENT.CENTER:
        element.set("alignment", "center")
    elif paragraph.alignment == WD_PARAGRAPH_ALIGNMENT.RIGHT:
        element.set("alignment", "right")
    elif paragraph.alignment == WD_PARAGRAPH_ALIGNMENT.JUSTIFY:
        element.set("alignment", "justify")
    else:
        element.set("alignment", "unknown")

def docx_to_xml(docx_file, xml_file):
    doc = Document(docx_file)
    root = ET.Element("document")
    is_table = False

    # Initialize position variables
    current_left = 0
    current_top = 0

    # Initialize variables to track headers and footers
    header_text = ""
    footer_text = ""

    for paragraph in doc.paragraphs:
        element = ET.Element("text")

        # Handle headings
        if paragraph.style.name.startswith("Heading"):
            element.tag = "header"

        # Handle numbered lists
        if re.match(r'^\d+\.', paragraph.text):
            element.tag = "list_item"

        # Handle links to sources (assuming URLs start with "http" or "www")
        if re.search(r'http|www', paragraph.text):
            element.tag = "link"

        # Set style attributes including typographic alignment
        style_to_attributes(paragraph, element)

        # Extract and add the text content
        element.text = paragraph.text

        # Capture and add the indentation (in points) as an attribute if defined
        if paragraph.paragraph_format.left_indent is not None:
            element.set("indent", str(paragraph.paragraph_format.left_indent.pt))

        # Check if we're inside a table
        if is_table:
            # Add the element to the last cell of the current row
            current_row[-1].append(element)
        else:
            # Add the element to the root
            root.append(element)

        # Check for headers and footers
        if paragraph.alignment == WD_PARAGRAPH_ALIGNMENT.CENTER:
            header_text += paragraph.text + "\n"
        elif paragraph.alignment == WD_PARAGRAPH_ALIGNMENT.RIGHT:
            footer_text += paragraph.text + "\n"

    # Handle tables
    for table in doc.tables:
        is_table = True
        table_element = ET.Element("table")
        for row in table.rows:
            current_row = []
            row_element = ET.SubElement(table_element, "row")
            for cell in row.cells:
                cell_element = ET.SubElement(row_element, "cell")
                current_row.append(cell_element)
                for paragraph in cell.paragraphs:
                    element = ET.Element("text")

                    # Set style attributes for text in table cells
                    style_to_attributes(paragraph, element)

                    # Extract and add the text content
                    element.text = paragraph.text

                    # Add the element to the cell
                    cell_element.append(element)

        # Add the table element to the root
        root.append(table_element)

    # Handle pictures
    for shape in doc.inline_shapes:
        if shape.type == 3:  # Check if it's a picture
            picture_element = ET.Element("picture")

            # Extract and add the size attributes
            picture_element.set("width", str(shape.width.pt))
            picture_element.set("height", str(shape.height.pt))

            # Calculate and add the position attributes
            picture_element.set("left", str(current_left))
            picture_element.set("top", str(current_top))

            # Add the picture element to the root
            root.append(picture_element)

        # Update the position variables for the next inline shape
        current_left += shape.width.pt
        current_top += shape.height.pt

    # Note headers and footers in the XML structure
    if header_text:
        header_element = ET.Element("header")
        header_element.text = header_text
        root.append(header_element)

    if footer_text:
        footer_element = ET.Element("footer")
        footer_element.text = footer_text
        root.append(footer_element)

    # Save the XML file
    tree = ET.ElementTree(root)
    tree.write(xml_file, encoding="utf-8", xml_declaration=True)

if __name__ == "__main__":
    docx_file = "output.docx"  # Replace with your Russian DOCX file path
    xml_file = "output8.xml"  # Output XML file path

    # Convert Russian DOCX to XML while preserving styles, indentation, tables, and typographic alignment
    docx_to_xml(docx_file, xml_file)
```

```python
import xml.etree.ElementTree as ET
from docx import Document
import PyPDF2  # Add PyPDF2 library
from reportlab.lib.pagesizes import letter
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.platypus.flowables import PageBreak
from reportlab.lib.styles import getSampleStyleSheet
import re
from PyPDF2 import PdfReader  # Use PdfReader from PyPDF2

# Helper function to convert DOCX paragraph styles to XML element attributes
def style_to_attributes(paragraph, element):
    run = paragraph.runs[0] if paragraph.runs else None

    # Font style
    if run and run.font.name:
        element.set("font", run.font.name)
    else:
        element.set("font", "Arial")  # Default font if not specified

    # Font size
    if run and run.font.size:
        element.set("font_size", str(run.font.size.pt))
    else:
        element.set("font_size", "12")  # Default font size if not specified

    # Bold, italic, and underline
    if run:
        element.set("bold", "true" if run.bold else "false")
        element.set("italic", "true" if run.italic else "false")
        element.set("underline", "true" if run.underline else "false")
    else:
        element.set("bold", "false")
        element.set("italic", "false")
        element.set("underline", "false")

    # Typographic alignment
    if paragraph.alignment == WD_PARAGRAPH_ALIGNMENT.LEFT:
        element.set("alignment", "left")
    elif paragraph.alignment == WD_PARAGRAPH_ALIGNMENT.CENTER:
        element.set("alignment", "center")
    elif paragraph.alignment == WD_PARAGRAPH_ALIGNMENT.RIGHT:
        element.set("alignment", "right")
    elif paragraph.alignment == WD_PARAGRAPH_ALIGNMENT.JUSTIFY:
        element.set("alignment", "justify")
    else:
        element.set("alignment", "unknown")

    # Leading (line spacing)
    if paragraph.paragraph_format.line_spacing is not None:
        # Convert line spacing value from lines to points (1 line = 12 points)
        leading_value = paragraph.paragraph_format.line_spacing * 12
        element.set("leading", str(leading_value))
    else:
        element.set("leading", "normal")  # Default leading if not specified

def generate_pdf(docx_file, pdf_file):
    # Load the DOCX file
    doc = Document(docx_file)

    # Initialize PDF document
    doc_pdf = SimpleDocTemplate(pdf_file, pagesize=letter)
    elements = []
    styles = getSampleStyleSheet()

    # Initialize position variables
    current_left = 0
    current_top = 0

    # Initialize variables to track headers and footers
    header_text = ""
    footer_text = ""

    for paragraph in doc.paragraphs:
        # Check for the page break marker and insert a PageBreak
        if "PAGE_BREAK" in paragraph.text:
            elements.append(PageBreak())
            continue

        # Create a paragraph element in the PDF
        p = Paragraph(paragraph.text, styles["Normal"])
        elements.append(p)

    # Build the PDF
    doc_pdf.build(elements)

def count_lines_on_pages(pdf_file):
    # Count lines on each page of the PDF using PyPDF2 PdfReader
    line_counts = []

    pdf_reader = PdfReader(open(pdf_file, 'rb'))
    
    for page_num in range(len(pdf_reader.pages)):
        page_text = pdf_reader.pages[page_num].extract_text()
        line_count = len([line for line in page_text.splitlines() if line.strip()])
        line_counts.append(line_count)

    return line_counts

if __name__ == "__main__":
    docx_file = "output.docx"  # Replace with your DOCX file path
    pdf_file = "output.pdf"  # Output PDF file path

    # Generate PDF from DOCX
    generate_pdf(docx_file, pdf_file)

    # Count lines on each page
    line_counts = count_lines_on_pages(pdf_file)

    # Create an XML structure to store line counts
    root = ET.Element("document")
    for count in line_counts:
        page_element = ET.Element("page")
        page_element.set("lines", str(count))
        root.append(page_element)

    # Save the XML file
    xml_file = "output.xml"
    tree = ET.ElementTree(root)
    tree.write(xml_file, encoding="utf-8", xml_declaration=True)
```
